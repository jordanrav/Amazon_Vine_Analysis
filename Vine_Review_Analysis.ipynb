```python
import pandas as pd
```

```python
df = pd.read_csv("vine_table.csv")
```

```python
df.head()
```

```python
tvotes_df=df[df["total_votes"] > 20]
tvotes_df.head()
```

```python
hvotes_df = tvotes_df[(tvotes_df['helpful_votes']/tvotes_df["total_votes"]) >= 0.5]
hvotes_df.head()
```

```python
yver_df = hvotes_df[hvotes_df["vine"]== 'Y']
yver_df.head()
```

```python
nver_df = hvotes_df[hvotes_df["vine"]== 'N']
nver_df.head()
```

```python
five_star_yver = yver_df['star_rating'].value_counts()[5]
tot_yver = yver_df['star_rating'].count()
five_star_nver = nver_df['star_rating'].value_counts()[5]
tot_nver = nver_df['star_rating'].count()

per_yver = five_star_yver/tot_yver
per_nver = five_star_nver/tot_nver
```

```python
print(f"the total number of five star reviews from the vine program are {five_star_yver} and the percentage is {round(per_yver , 2)*100}%")
```

```python
print(f"the total number of five star reviews from the vine program are {five_star_nver} and the percentage is {round(per_nver , 2)*100}%")
```
